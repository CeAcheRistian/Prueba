```python
import pandas as pd #Importamos la librería de pandas
```

```python
df = pd.Series([1,2,3]) #Las Series de pandas son arreglos unidimensionales que pueden almacenar cualquier tipo de dato
df #Así se imprime un objeto tipo panda, al ser un arreglo, se le da un indice automáticamente
#Los llamamos df por Data Frame
```

```python
df2 = pd.Series([1,2,3,4,5], index = ["0","1","2","3","4"]) #Dicho índice puede ser modificado, pero solo permite cadenas de números ordenados y enteros ordenados
df2["1"] #Se accede a un elemento en específico de la lista
```

```python
df3 = pd.Series({"uno": 1, "dos": 2, "tres":3, "cuatro": 4}) #También se pueden diccionarios, en vez de listas, la llave es el indice, el valor es el valor pero de la lista
df3
```

```python
type(df3) #El tipo de dato de un series.Series
```

```python
type(df3["uno"]) #y el tipo de dato de almacena el valor del diccionario no es un entero tal cual, sino un numpy.int64
```

```python
df4 = pd.DataFrame({ #Creamos un diccionario que se imprimirá como tabla. Es más usual ver una llamada a DataFrame
    "nombre":["Christian", "Jesus", "Galileo"],
    "edad":[1,24, 7],
    "color":["negro", "amarillo", ""]
})# Las columnas se llaman como las llaves y sus datos se van repartiendo
df4
```

```python
df4.index #Nos dice en donde empieza el indice, donde para y la cuenta
```

```python
df4.columns #Nos da el nombre de las columnas y su tipo
```

```python
l1 = [1,2,3,4,5,6]
l2 = [0,1,2,3,4,5]
df5 = pd.DataFrame(data= l1, index = l2)
df5 #Es posible unir los listas, una serian los valores y la otra los indices de los valores.
```

```python
l1 = {"nombre": [True, False], "color": [2,3] }
l2 = [0,1]
df6 = pd.DataFrame(data= l1, index = l2)
df6
```

```python
df7 = pd.DataFrame({
    "nombre":["Christian", "Jesus", "Galileo", "Alejandro", "Liliana", "Skiper"],
    "edad":[1,24, 7,45, 12, 45],
    "color":["negro", "amarillo", "verde", "negro", "violeta", "naranja"]
}) #Existen varias formas de filtrar la información
df7[df7["edad"]<18] #Para los tipo enteros, con < > 
df7[df7["nombre"]=="Skiper"] #Para las cadenas con ==
df_filtrado = df7[df7["color"]== "negro"] #Es posible almacenar en variables los resultados de un filtrado de un dataframe
```

```python
df_filtrado
```

```python
df_ordenado= df7.sort_values("edad") #Con sort_values ordenamos un DF de enteros
```

```python
df_ordenado
```

```python
df_nuevo = df_ordenado["edad"] #Seleccionamos solo la columna edad
df_nuevo.min() #Y como son enteros, podemos llamar a las funciones de minimo, maximo y promedio
df_nuevo.max()
df_nuevo.mean()
```

```python
for i in df7: #Podemos hacer ciclos for de los datos del DF. Por defecto nos dará el nombre de las columnas
    print(i)
```
